```python
# Load dataset and preprocess audio files
def load_dataset(data_dir, sample_rate=16000, noise=False):
    commands = os.listdir(data_dir)
    commands = [cmd for cmd in commands if os.path.isdir(os.path.join(data_dir, cmd))]
    audio_data, labels = [], []

    for idx, command in enumerate(commands):
        command_dir = os.path.join(data_dir, command)
        for file in os.listdir(command_dir):
            file_path = os.path.join(command_dir, file)
            try:
                # Load only audio files
                if file_path.endswith('.wav'):  # Ensure it's an audio file
                    audio, sr = librosa.load(file_path, sr=sample_rate)
                    if noise:
                        audio = add_noise(audio)
                    audio_data.append(audio)
                    labels.append(idx)
            except Exception as e:
                print(f"Skipping file {file_path}: {e}")

    return np.array(audio_data, dtype=object), np.array(labels)
```
